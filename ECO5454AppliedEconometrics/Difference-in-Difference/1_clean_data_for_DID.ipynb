## import the affected county to identify which county is affected

```python
pd_emp_list3=pd.read_csv('louisana_clean_emp.csv')

affected=pd.read_csv('affected county in Katrina.csv')
```

```python
affected_name=affected['County FIPS'].unique().tolist()
```

```python
affected_name
```

```python
affected_fips_list=[]
for i in range(len(affected)):
    fips=int(eval(affected['County FIPS'][i]))
    affected_fips_list.append(fips)
    
    
```

```python
pd_emp_list3['affected_county']=''

for i in range(len(pd_emp_list3)):
    area_fipsi=pd_emp_list3['area_fips'][i]
    #print(area_fipsi)
    if area_fipsi in affected_fips_list:
        pd_emp_list3['affected_county'][i]=1
        
    else: 
        pd_emp_list3['affected_county'][i]=0
```

```python
#add the column of preshock and post chock 
pd_emp_list3['after_2005Q3']=''
#  shock happens at 2005 quarter 3. so 1 means after shock, 0 means before shock.
for i in range(len(pd_emp_list3)):
    
    yeari= pd_emp_list3['year'][i]
    qtri=pd_emp_list3['qtr'][i]
   # print(type(yeari))
    if yeari ==2003 or yeari ==2004:
        pd_emp_list3.loc[i,'after_2005Q3']=0
    if yeari ==2006 or yeari ==2007:
        #print(yeari)
        pd_emp_list3.loc[i,'after_2005Q3']=1
    if yeari==2005 and qtri==4:
        pd_emp_list3.loc[i,'after_2005Q3']=1
    if yeari==2005 and ((qtri==1)or(qtri==2)or(qtri==3)):
        pd_emp_list3.loc[i,'after_2005Q3']=0
        
        

```

```python
pd_emp_list3
```

```python
pd_emp_list3['affected_county_after_shock']=''

pd_emp_list3['affected_county_after_shock']=pd_emp_list3['affected_county']*pd_emp_list3['after_2005Q3']
```
